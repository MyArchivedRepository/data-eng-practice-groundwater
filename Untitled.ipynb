```python
import pandas as pd
import requests
from bs4 import BeautifulSoup 

def get_all_data() :
     
    def get_data(i_pages):
        url = f"http://app.dgr.go.th/newpasutara/xml/Krabi.files/show_xls.php?ddlGeo1={i_pages}&ddlProvince1=&ddlAmphur1="
        data = pd.read_html(url)[0]
        return data
    total_data = [get_data(i) for i in range(1,4)]    
    result = pd.concat(total_data,ignore_index=True).to_csv("groundwater_data.csv", index=False, encoding='utf-8-sig')


```
